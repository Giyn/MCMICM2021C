# 数据预处理

## 相关设置

```python
from IPython.core.interactiveshell import InteractiveShell

import pandas as pd
import numpy as np

InteractiveShell.ast_node_interactivity = "all"  # 一个cell显示多输出

pd.set_option('display.max_rows', 100)  # 设置最大显示100行
pd.set_option('display.max_columns', 100)  # 设置最大显示100列
```

```python
Images_by_GlobalID = pd.read_excel("data/raw_data/2021MCM_ProblemC_Images_by_GlobalID.xlsx")
DataSet = pd.read_excel("data/raw_data/2021MCMProblemC_DataSet.xlsx")
```

```python
try:
    Images_by_GlobalID.drop(labels='Unnamed: 3', axis=1, inplace=True)  # 去除最后一列的无用列
except KeyError:
    pass
Images_by_GlobalID.head()
DataSet.head()
```

## 化简文件名

```python
Images_by_GlobalID['FileName'] = Images_by_GlobalID['FileName'].map(lambda x: '{}.png'.format(x.split('_')[0]))
Images_by_GlobalID.head()
```

```python
Images_by_GlobalID.info()
```

## 制作 GlobalID-FileName 映射表

```python
GlobalIDList = Images_by_GlobalID['GlobalID'].tolist()
FileNameList = Images_by_GlobalID['FileName'].tolist()
GlobalID_To_FileName_Map = dict(zip(GlobalIDList, FileNameList))
```

```python
# print(GlobalID_To_FileName_Map)
```

```python
DataSet['FileName'] = DataSet['GlobalID'].map(GlobalID_To_FileName_Map)
```

```python
DataSet.head()
```

```python
DataSet.to_csv("data/processed_data.csv", index=False, encoding='utf_8_sig')
```
